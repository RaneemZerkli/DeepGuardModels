```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Generate multiple sentences using beam search
max_length = 150  # Adjust the desired maximum length for the generated text
num_sentences = 3  # Number of sentences to generate
num_beams = 5  # Number of beams to search for each sentence

generated_text = model.generate(
    max_length=max_length,
    num_return_sequences=num_sentences,
    no_repeat_ngram_size=2,
    top_k=50,
    top_p=0.95,
    temperature=0.7,
    num_beams=num_beams,
)

# Decode and print the generated text
for i, text in enumerate(generated_text):
    decoded_text = tokenizer.decode(text, skip_special_tokens=True)
    print(f"Sentence {i + 1}: {decoded_text}\n")

```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can choose a specific GPT-2 variant (small, medium, large, etc.)
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name, padding_side='left')

# Generate and save sentences incrementally
max_length = 150  # Adjust the desired maximum length for the generated text
num_sentences = 10_000  # Total number of sentences to generate
batch_size = 500  # Number of sentences to generate in each batch
output_file = "generated_sentences.txt"

# Initialize an empty list to store generated sentences
generated_sentences = []

# Generate sentences in batches
for batch_start in range(0, num_sentences, batch_size):
    batch_end = min(batch_start + batch_size, num_sentences)

    # Generate text for the current batch
    output = model.generate(
        max_length=max_length,
        num_return_sequences=batch_end - batch_start,
        no_repeat_ngram_size=2,
        top_k=50,
        do_sample=True,
        pad_token_id=50256,
    )

    # Decode and store the generated text
    for i, text in enumerate(output):
        decoded_text = tokenizer.decode(text, skip_special_tokens=True)
        generated_sentences.append(f"Sentence {batch_start + i + 1}: {decoded_text}")

    # Save generated sentences incrementally
    with open(output_file, "a", encoding="utf-8") as file:
        for sentence in generated_sentences:
            file.write(sentence + "\n")

    # Clear the list for the next batch
    generated_sentences = []

print("Generated and saved all sentences to", output_file)
```
