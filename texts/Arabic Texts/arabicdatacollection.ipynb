```python
import transformers
from arabert import ArabertPreprocessor

# Load the pre-trained Arabic GPT-2 model and tokenizer
model_name = "aubmindlab/aragpt2-base"
preprocessor = ArabertPreprocessor(model_name=model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Set the number of sentences to generate
num_sentences = 4000

# Initialize a list to store the generated sentences


# Generate the text in batches
for i in range(0, num_sentences, 100):
    # Preprocess the input text
    input_text = ""
    input_text_preprocessed = preprocessor.preprocess(input_text)

    # Generate the text
    output = model.generate(max_length=22, num_return_sequences=100, no_repeat_ngram_size=2, top_k=50, do_sample=True, pad_token_id=50256)

    # Decode and append the generated text to the list
    for j, sample_output in enumerate(output):
        generated_text = tokenizer.decode(sample_output, skip_special_tokens=True)
        generated_sentences.append(generated_text)


```

```python
df=pd.DataFrame({'Inputs':generated_sentences,'Label':0})
```
